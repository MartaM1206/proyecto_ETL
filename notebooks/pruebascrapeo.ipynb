```python
import requests
from bs4 import BeautifulSoup

# URL de la página web que deseas scrappear
url_hoteles = "https://all.accor.com/ssr/app/ibis/hotels/madrid-spain/index.es.shtml?compositions=1&stayplus=false&snu=false&hideWDR=false&productCode=null&accessibleRooms=false&hideHotelDetails=false"


# Realiza una solicitud GET a la página web
response = requests.get(url_hoteles)
html_content = response.content

# Analiza el contenido HTML de la página
soup = BeautifulSoup(html_content, 'html.parser')

# Encuentra todos los bloques de hotel con la clase hotelblock__content
bloques_hoteles = soup.find_all("div", {"class": "hotelblock__content"})

# Itera sobre los bloques encontrados y extrae la información
for i, bloque in enumerate(bloques_hoteles, start=1):
    # Acceder al nombre del hotel
    nombre_hotel = bloque.select_one('.title__link').get_text(strip=True)
    
    # Acceder al número de estrellas del hotel
    estrellas = len(bloque.select('.rating__star'))
    
    # Acceder al precio del hotel
    precio_elemento = bloque.find('span.booking-price__number mcp-price-number')
    if precio_elemento:
        precio = precio_elemento.get_text(strip=True)
    else:
        precio = "Precio no encontrado"
    
    print(f'Hotel {i}:')
    print('Nombre:', nombre_hotel)
    print('Estrellas:', estrellas)
    print('Precio:', precio)
    print('---' * 10)
```
